```python
import matplotlib.pyplot as plt
import numpy as np

# list T_Real
T_Real = np.array([22.3, 22.7,23.0 , 23.1 , 23.3 , 24.7 , 24.6 , 23.8 , 21.1, 11, 21.7, 25.2, 25 ])

#list T_LM35
T_LM35 = np.array([34.23,37.03 , 31.72 , 30.89, 32.23 , 34.18 ,42.97 ,36.43 , 35.82,32.29, 45.88, 31.23, 31.25])

#list T_DHT
T_DHT = np.array([25.44 , 24.48, 25.44 , 25.38 , 26.44 , 27.66 , 27.36, 27.26 , 15.82 , 12.16, 24.58, 29.54, 29.60 ])

#T_LM35 trend

a1 , b1 = np.polyfit(T_LM35, T_Real, 1)

print(f"Γραμμική εξίσωση LM35:")
print(f"T_Real= {a1: .3f} * T_LM35 + {b1: .3f}")

trendline_LM35 = a1 * T_LM35 + b1
idx1 = np.argsort(T_LM35)
#T_DHT trend

a2, b2 = np.polyfit(T_DHT, T_Real, 1)

print (f"\nΓραμμική εξίσωση DHT:")
print(f"T_Real = {a2: .3f} * T_DHT + {b2: .3f}")

trendline_DHT = a2 * T_DHT + b2
idx2 = np.argsort(T_DHT)
plt.figure(figsize=(12,5))

#Γράφημα LM35

plt.subplot(1, 2 , 1)
plt.plot(T_LM35, T_Real, label= 'Δεδομένα')
plt.plot(T_LM35[idx1], trendline_LM35[idx1], label = 'Trendline')

plt.xlabel("T_LM35")
plt.ylabel("T_Real")
plt.title("LM35 vs Πραγματική θερμοκρασία")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("LM35_calibration.png", dpi=300, bbox_inches="tight")
plt.show()

#Γράφημα DHT

plt.subplot(1,2,2)

plt.scatter(T_DHT,T_Real,label = 'Δεδομένα')
plt.plot(T_DHT[idx2], trendline_DHT [idx2], label = 'Trendline')

plt.xlabel("T_DHT")
plt.ylabel("T_Real")
plt.title("DHT vs Πραγματική Θερμοκρασία")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("DHT11_calibration.png",dpi=300, bbox_inches="tight")
plt.show()

from google.colab import files
files.download("LM35_calibration.png")
files.download("DHT11_calibration.png")



```
